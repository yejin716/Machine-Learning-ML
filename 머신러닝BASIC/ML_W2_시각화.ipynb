```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
path = '../../data/data자료/Advertising.csv'
pd.read_csv(path)
```

```python
ad = pd.read_csv(path)
ad.head()
```

```python
ad.columns
```

```python
#데이터 생성 

tv = ad['TV']
radio = ad['Radio']
news = ad['Newspaper']
sales = ad['Sales']


#plot그리기 
fig ,ax = plt.subplots()

ax.plot(tv, linewidth=1.0)
ax.plot(radio, linewidth=1.0)
ax.plot(news, linewidth=1.0)
ax.plot(sales, linewidth=1.0)

plt.grid()
plt.show()
```

```python
#scatter()

#데이터 생성 
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24) #정규분포(평균, 표준편차, 생성할 개수)
y = 4 + np.random.normal(0, 2, len(x))

#size / color
sizes = np.random.uniform(15, 80, len(x)) #uniform () 동일한 확률로15~80개 x의 크기만큼 
colors = np.random.uniform(15, 80, len(x))
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors)
#plot

ax.set(xlim=(0,8), xticks=np.arange(1,8),
       ylim=(0,8), yticks=np.arange(1,8))
plt.grid()
plt.show()
```

```python
#TV 광고예산과 판매량과 관계

fig, ax = plt.subplots()

ax.scatter(tv, sales, s=5, c='blue')

ax.set(xlabel=('TV'),
        ylabel=('Sales'))

plt.grid()
plt.show()
```

```python
#Radio 광고예산과 판매량과 관계

fig, ax = plt.subplots()

ax.scatter(radio, sales, s=5, c='orange')

ax.set(xlabel=('Radio'),
        ylabel=('Sales'))

plt.grid()
plt.show()
```

```python
#news 광고예산과 판매량과 관계

fig, ax = plt.subplots()

ax.scatter(news, sales, s=5, c='green')

ax.set(xlabel=('News'),
        ylabel=('Sales'))

plt.grid()
plt.show()
```

```python
#히스토그램 hist()\
    
#데이터 생성
np.random.seed(2)

x = 4 + np.random.normal(0, 1.5, 200)

fig, ax = plt.subplots()

ax.hist(x, bins=8, linewidth=0.5, edgecolor='white')

ax.set(xlim=(0,8), xticks=np.arange(1,8),
       ylim=(0,60), yticks=np.linspace(0, 60, 9))
plt.grid()
plt.show()
```

```python
# TV 광고예산 분포 

fig, ax = plt.subplots()

ax.hist(tv, bins=8, linewidth=0.5, edgecolor='white')

ax.set(xlabel=('TV'),
       ylabel=('Number of Markets'))

plt.grid()
plt.show()
```

```python
# news 광고예산 분포 

fig, ax = plt.subplots()

ax.hist(news, bins=8, linewidth=0.5, edgecolor='white')

ax.set(xlabel=('News'),
       ylabel=('Number of Markets'))

plt.grid()
plt.show()
```

```python
#boxplot ()

#데이터 생성
np.random.seed(10)
d = np.random.normal((3,5,4),(1.25,1.00,1.25),(100,3))

#데이터 총 3개 
#평균 3, 표준편차 1.25, 100개 생성 

#plot
fig,ax = plt.subplots()
ax.boxplot(d)

plt.grid()
plt.show()
```

```python
#3가지 광고매체 비교 
d = [tv, radio, news]

#plot
fig, ax = plt.subplots()
ax.boxplot(d)

ax.set(xlabel=('Media'),
       ylabel = ('Budget'))

plt.xticks([1,2,3], ['TV','Radio','News'])

plt.grid()
plt.show()
```

```python
import seaborn as sns 

#relplot()
sns.relplot(x=tv, y=sales)
plt.show()
```

```python
#PairGrid()

g = sns.PairGrid(ad) #여러개 변수를 하나로 표현 
g.map_diag(sns.histplot) #대각선은 histogram이 나오게함
g.map_offdiag(sns.scatterplot) # 그 외 scatter가 나옴

plt.show()
```

```python
#regression line


g = sns.PairGrid(ad, y_vars=['Sales'], x_vars=['TV', 'Radio', 'Newspaper'])
g.map(sns.regplot, color='0.4')
plt.show()
```
