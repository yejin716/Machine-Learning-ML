```python
#재표본 추출 방법(resampling)

import numpy as np 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

from sklearn.model_selection import KFold #k-fold cross validation 
from sklearn.utils import resample # bootstrap

#임의로 데이터를 섞기 위해(random shuffling)
# >> 편향 방지 위함 
from sklearn.utils import shuffle

from sklearn.datasets import load_iris
```

K-Fold cross validation

```python
#간단 예제 

x = ['a','b','c','d']

kf = KFold(n_splits=2) # 2개로 나누기 

for train, test in kf.split(x):
    print("%s %s" %(train, test))
    print()
    print(x[train[0]], x[train[1]], x[test[0]], x[test[1]])
    print()
    
```

```python
print(x[train[0]], x[train[1]])
```

```python
print(x[test[0]], x[test[1]])
```

iris data 활용 , KFold cross validation 

```python
x_iris, y_iris = load_iris(return_X_y=True)
```

```python
print(x_iris)
print()
print(y_iris)
```

```python
# 5 fold cross validation 

kf = KFold(n_splits=5)
idx = 1 

for train, val in kf.split(x_iris):
    print("Fold:", idx)
    
    #k-fold 결과 확인 
    print("%s %s" %(train,val))
    print()
    print(y_iris[train])
    print()
    
    print(y_iris[val])
    print()
    
    print('--' * 100)
    idx += 1 
```

```python
# 5 fold cross validation 

kf = KFold(n_splits=5)
idx = 1 

for train, val in kf.split(x_iris):
    print("Fold:", idx)
    
    #LDA 훈련 
    lda = LinearDiscriminantAnalysis().fit(x_iris[train], y_iris[train])
    
    #클래스 예측 
    print("True label:", y_iris[val])
    print("Pred label:", lda.predict(x_iris[val]))
    
    #accuracy 
    print("Accuracy:", lda.score(x_iris[val], y_iris[val]))
    
    
    print('--' * 100)
    idx += 1 
```

bootstrap (간단한 예제)

```python
x = np.array([[1.,0.], [2.,1.],[0.,0.]])
y = np.array([0,1,2])

print(x)
print(y)
```

```python
#재표본 추출 (랜덤)
x, y = resample(x, y)
print(x)
print(y)
```
