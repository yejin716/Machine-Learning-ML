```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
print(iris)
print()
print(iris.data)
print()
print(iris.target)
print()
print(iris.feature_names)
print()
print(iris.target_names)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head(3)
```

```python
df['label'] = iris.target
df.head(3)
```

```python
#x, y >> train / teset
x = iris.data
y = df['label']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
#머신러닝 알고리즘 선택 >> 모델링

#의사결정나무 (decissionTreeClassifier)

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()

#훈련용 데이터 >> 모델 구축
dt.fit(x_train, y_train)
```

```python
#모델 성능평가 (정확도 평가)

from sklearn.metrics import accuracy_score

#예측 (평가용데이터로)
y_pred = dt.predict(x_test)
accuracy_score(y_test, y_pred)
```

라벨인코딩 실습

```python
from sklearn.preprocessing import LabelEncoder

items = ['TV', '냉장고', '전자레인지', '컴퓨터', '아이패드', '노트북', '모니터']

# 라벨인코더 모듈 불러오기
le = LabelEncoder()
# 라벨인코더 데이터(items)훈련 >> 변환(숫자 인코딩)
le.fit(items)

labels = le.transform(items)
labels
```

```python
#원상 복구
#Decode

origin = le.inverse_transform(labels)
origin
```
