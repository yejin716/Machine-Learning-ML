분석 데이터 셋 준비
- X : 독립변수(설명변수), y는 종속변수(목표변수)

```python
df['family'] =  df[['SibSp', 'Parch']]
```

```python
df.columns
```

```python
df = [['Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Embarked']]
```

```python
x = df[['Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Embarked']]
y = df['Survived']
```

```python
#데이터 분할

x_train, x_test, y_train, y_test =\
train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#모델링
# 의사결정나무 모델 가져오기

dt = DecisionTreeClassifier(random_state=42)

#학습 >> 모델생성(dt model)
dt.fit(x_train, y_train) #훈련
```

```python
#학습이 완료된 dt model 가지고 test data로 예측 수행
y_pred = dt.predict(x_test)
```

```python
#우리의 목표 : 에러 최소화
# >> y_pred과 y_real(y_test)의 차이가 0에 가깝게 만드는 것
```

```python
y_pred #예측값
```

```python
print(y_test) #실제값
```

```python
#모델 성능비교 - 정확도 측정

from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)
acc
```
